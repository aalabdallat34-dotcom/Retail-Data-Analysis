```python
RetailStore_copy=RetailStore.copy()
```

```python
mask = (
    RetailStore_copy['Price Per Unit'].isnull() &
    RetailStore_copy['Quantity'].notnull() &
    RetailStore_copy['Total Spent'].notnull()
)

RetailStore_copy.loc[mask, 'Price Per Unit'] = (
    RetailStore_copy.loc[mask, 'Total Spent'] /
    RetailStore_copy.loc[mask, 'Quantity']
)
```

```python
RetailStore_copy.isnull().sum()
```

```python
item_price_map = (
    RetailStore_copy
    .dropna(subset=['Item', 'Category', 'Price Per Unit'])
    .groupby(['Category', 'Price Per Unit'])['Item']
    .nunique()
)

item_price_map[item_price_map > 1]
```

```python
item_lookup = (
    RetailStore_copy[RetailStore_copy['Item'].notnull()]
    [['Category', 'Price Per Unit', 'Item']]
    .drop_duplicates()
)

RetailStore_copy = RetailStore_copy.merge(
    item_lookup,
    on=['Category', 'Price Per Unit'],
    how='left',
    suffixes=('', '_new')
)

RetailStore_copy['Item'] = (
    RetailStore_copy['Item']
    .fillna(RetailStore_copy['Item_new'])
)

RetailStore_copy.drop(columns=['Item_new'], inplace=True)
```

```python
complete_rows = RetailStore_copy.dropna(
    subset=['Price Per Unit', 'Quantity', 'Total Spent']
).copy()

complete_rows['Expected Total'] = (
    complete_rows['Price Per Unit'] *
    complete_rows['Quantity']
)

inconsistent_rows = complete_rows[
    complete_rows['Expected Total'] != complete_rows['Total Spent']
]

print("Complete rows:", len(complete_rows))
print("Inconsistent rows:", len(inconsistent_rows))
print("Inconsistency percentage:",
      len(inconsistent_rows) / len(complete_rows) * 100)
```
